```python
def total(name, start_rate, end_rate, t1, t3, save_space):
    # html구조가 full인지 simple인지 확인하기
    driver.execute_script("window.scrollTo(0, 600);")
    sleep(t3)
    try:
        rate1 = driver.find_element_by_xpath('//*[@id="taplc_location_detail_reviews_card_0"]/div[3]/div/ul/li[1]/span[1]')
        mode = "full"
    except:
        try:
            rate1 = driver.find_element_by_xpath('//*[@id="taplc_detail_filters_acr_responsive_0"]/div/div[1]/div/div[2]/div[1]/div/div[2]/div/div[1]/label')
            mode = "simple"
        except:
            print(name + " cannot classify. 수동으로 체크 필요")

    # 초기화
    head_list2 = []
    text_list2 = []
    time_list2 = []
    rate_list = []
    
    # full mode라면
    if mode == "full":
        for k in range(start_rate, end_rate+1):
            # 특정 평점 클릭하기
            try:
                clickon_rate1(k,t3)
            except:
                print(name + "'s rate " + str(k) + " is not cralwed. 수동으로 추가 크롤링 필요")
                continue
            
            # 페이지 넘기며 크롤링하기
            page1(name, k, t1, head_list2, text_list2, time_list2, rate_list)
            
            # 저장하기
            temp_crawl = {'head':head_list2, 'review':text_list2, 'date':time_list2, 'rate':rate_list}
            save(name, save_space, temp_crawl)
            
    # simple mode라면
    elif mode == "simple":
        for k in range(start_rate, end_rate+1):
            # 특정 평점 클릭하기
            try:
                clickon_rate2(k,t3)
            except:
                print(name + "'s rate " + str(k) + " is not cralwed. 수동으로 추가 크롤링 필요")
                continue
            
            # 페이지 넘기며 크롤링하기
            page2(name, k ,t1, head_list2, text_list2, time_list2, rate_list)
            
            # 다음 평점을 위해 현재 평점 클릭 해제하기
            try:
                clickoff_rate(k, t3)
            except:
                try:
                    clickoff_rate(k, t3)
                except:
                    print(name + "'s rate " + str(k) + " is not clicked off.... disastrous! ㅠㅠ")
                    continue
                      
            # 저장하기
            temp_crawl = {'head':head_list2, 'review':text_list2, 'date':time_list2, 'rate':rate_list}
            save(name, save_space, temp_crawl)
```

```python
# 2차 카테고리에서 '최고의 리뷰'를 보인 상위 30여개 크롤링해오기
# country: 국가명
# m : 1차 카테고리 number
# start: 2차 카테고리 중 몇번째에서 시작할 것인가
# t0: 특정 관광지 클릭 후 기다리는 시간
# t1: 페이지 넘긴 후 기다리는 시간
# t3: 스크롤 후 기다리는 시간

def best_reviews(country, m, start_place, t0, t1, t3, save_space):
    #driver.execute_script("window.scrollTo(0, 1200);")
    #sleep(1)
    candidate = driver.find_elements_by_link_text("더 알아보기")
    
    # 특정 여행지로 들어가기
    for t in range(start_place, len(candidate)+1):
        candidate[t].click()
        sleep(t0)            
        
        # 특정 여행지 탭으로 이동
        driver.switch_to.window(driver.window_handles[-1])
        
        # 해당 여행지 명 확인하기
        try:
            name = str(country) + str(m) + '_' + str(t) + '_' + driver.find_element_by_xpath('//*[@id="HEADING"]/div').text
        except:
            name = str(country) + str(m) + '_' + str(t) + '_' + driver.find_element_by_xpath('//*[@id="HEADING"]').text
            
        # 해당 여행지의 평가들 크롤링해오기
        total(name, start_rate, end_rate, t1, t3, save_space)
        
        # 크롤링 완료 후 탭 종료 및 이전 탭으로 이동하기
        driver.close()
        sleep(0.5)
        driver.switch_to.window(driver.window_handles[-1])
        sleep(0.5)
```
